```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import random

########################################
# 1. Druhy poƒç√≠taƒçov√© grafiky
########################################

display(HTML("<h2>Poƒç√≠taƒçov√° grafika ‚Äì Druhy</h2>"))
display(HTML("<b>Ot√°zka 1:</b> Jak√© m√°me hlavn√≠ druhy poƒç√≠taƒçov√© grafiky?"))

q1_options = [
    "3D grafika a 2D grafika",
    "Bitov√° grafika a analogov√° grafika",
    "Rastrov√° grafika a vektorov√° grafika"
]
# odstranili jsme description="Druhy:"
q1_radio = widgets.RadioButtons(options=q1_options, description='')
q1_button = widgets.Button(description="Ovƒõ≈ôit odpovƒõƒè", button_style="info")
q1_output = widgets.Output()

def check_q1(b):
    with q1_output:
        clear_output()
        if q1_radio.value == "Rastrov√° grafika a vektorov√° grafika":
            print("Spr√°vnƒõ!")
        else:
            print("Nespr√°vnƒõ, zkuste to znovu.")

q1_button.on_click(check_q1)
display(q1_radio, q1_button, q1_output)
########################################
# 2. Form√°tov√© p≈ôi≈ôazen√≠ (upraveno)
########################################

display(HTML("<h2>Form√°tov√© p≈ôi≈ôazen√≠</h2>"))
display(HTML("<b>Ot√°zka 2:</b> P≈ôi≈ôaƒète ka≈æd√Ω form√°t ke spr√°vn√©mu typu grafiky:"))

formats = [
    ("JPG/JPEG", "Rastrov√°"),
    ("PNG", "Rastrov√°"),
    ("GIF", "Rastrov√°"),
    ("TIFF", "Rastrov√°"),
    ("SVG", "Vektorov√°"),
    ("AI", "Vektorov√°"),
    ("EPS", "Vektorov√°"),
    ("PDF", "Vektorov√°")
]
# Prom√≠ch√°me po≈ôad√≠ form√°t≈Ø
random.shuffle(formats)

dropdowns = {}
boxes = []
for fmt, correct in formats:
    label = widgets.HTML(value=f"<b>{fmt}</b>")
    dd = widgets.Dropdown(
        options=["Vyberte", "Rastrov√°", "Vektorov√°"],
        value="Vyberte",
        description='',
        layout=widgets.Layout(width="150px")
    )
    dropdowns[fmt] = dd
    box = widgets.VBox([label, dd])
    boxes.append(box)

q2_button = widgets.Button(description="Ovƒõ≈ôit odpovƒõdi", button_style="info")
q2_output = widgets.Output()

def check_q2(b):
    with q2_output:
        clear_output()
        all_correct = True
        for fmt, dd in dropdowns.items():
            correct_answer = [c for f, c in formats if f == fmt][0]
            if dd.value == "Vyberte":
                print(f"{fmt}: Je≈°tƒõ jste nevybrali odpovƒõƒè.")
                all_correct = False
            elif dd.value != correct_answer:
                print(f"{fmt}: Nespr√°vnƒõ")
                all_correct = False
            else:
                print(f"{fmt}: Spr√°vnƒõ")
        if all_correct:
            print("V≈°echny odpovƒõdi jsou spr√°vn√©!")
        else:
            print("Zkuste to pros√≠m znovu.")

q2_button.on_click(check_q2)

grid = widgets.GridBox(
    children=boxes,
    layout=widgets.Layout(
        grid_template_columns="repeat(auto-fill, minmax(150px, 1fr))",
        grid_gap="10px"
    )
)
display(grid)
display(q2_button, q2_output)


########################################
# 3. V√Ωhody a nev√Ωhody grafiky
########################################

display(HTML("<h2>V√Ωhody a nev√Ωhody grafiky</h2>"))
display(HTML("<b>Ot√°zka 3:</b> Jak√° je nev√Ωhoda rastrov√© grafiky p≈ôi zvƒõt≈°en√≠?"))

q3_options = [
    "Obr√°zek z≈Øst√°v√° ostr√Ω",
    "Zvƒõt≈°en√≠ nem√° ≈æ√°dn√Ω vliv",
    "Obr√°zek se st√°v√° rozmazan√Ωm"
]
q3_radio = widgets.RadioButtons(options=q3_options, description="Nev√Ωhoda:")
q3_button = widgets.Button(description="Ovƒõ≈ôit odpovƒõƒè", button_style="info")
q3_output = widgets.Output()

def check_q3(b):
    with q3_output:
        clear_output()
        if q3_radio.value == "Obr√°zek se st√°v√° rozmazan√Ωm":
            print("Spr√°vnƒõ!")
        else:
            print("Nespr√°vnƒõ, zkuste to znovu.")

q3_button.on_click(check_q3)
display(q3_radio, q3_button, q3_output)

########################################
# 4. RGB ‚Äì co to znamen√°?
########################################

display(HTML("<h2>RGB</h2>"))
display(HTML("<b>Ot√°zka 4:</b> Co znamen√° zkratka RGB?"))

q4_text = widgets.Text(placeholder="Napi≈°te, co znamen√° RGB", description="")
q4_button = widgets.Button(description="Ovƒõ≈ôit odpovƒõƒè", button_style="info")
q4_output = widgets.Output()

def check_q4(b):
    with q4_output:
        clear_output()
        answer = q4_text.value.strip().lower()
        if answer in ["red green blue", "ƒçerven√° zelen√° modr√°", "ƒçerven√°, zelen√°, modr√°", "red, green, blue"]:
            print("Spr√°vnƒõ!")
        else:
            print("Nespr√°vnƒõ!")

q4_button.on_click(check_q4)
display(q4_text, q4_button, q4_output)
```

```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import matplotlib.pyplot as plt
import io
import base64
from IPython.display import Image

# Definice kategori√≠ a mo≈ænost√≠ s jejich bodov√Ωm ohodnocen√≠m
categories = {
    "P≈ô√≠prava videa": [
        {"type": "radio", "name": "D√©lka √∫vodu", "options": [
            ("Kr√°tk√Ω √∫vod", 500),
            ("Dlouh√Ω √∫vod", -1000),
            ("Bez √∫vodu", 0)
        ]},
        {"type": "checkbox", "name": "Toƒç√≠m se sc√©n√°≈ôem", "value": 300}
    ],
    "Kvalita produkce": [
        {"type": "radio", "name": "Technick√° kvalita", "options": [
            ("Kvalitn√≠ obraz, ale nekvalitn√≠ zvuk", -1500),
            ("Kvalitn√≠ zvuk, ale hor≈°√≠ obraz", 1000),
            ("Natoƒç√≠m video se zvukem dohromady na jednom za≈ô√≠zen√≠", -1500)
        ]}
    ],
    "Propagace": [
        {"type": "checkbox", "name": "Sd√≠len√≠ na soci√°ln√≠ch s√≠t√≠ch", "value": 2000},
        {"type": "radio", "name": "N√°hledov√Ω obr√°zek", "options": [
            ("Udƒõl√°m na YouTube pƒõknou n√°hledovku", 1500),
            ("N√°hledovku na YouTube dƒõlat nebudu", -500),
            ("Pou≈æiju generickou n√°hledovku", 100)
        ]},
        {"type": "checkbox", "name": "D√°m video do playlistu", "value": 300}
    ],
    "U≈æivatelsk√Ω z√°≈æitek": [
        {"type": "checkbox", "name": "P≈ôid√°m titulky", "value": 500},
        {"type": "checkbox", "name": "Pou≈æiji vhodnou hudbu v pozad√≠", "value": 800},
        {"type": "checkbox", "name": "P≈ô√≠li≈° mnoho reklam", "value": -2000}
    ],
    "Naƒçasov√°n√≠": [
        {"type": "radio", "name": "ƒåas vyd√°n√≠", "options": [
            ("Vyd√°n√≠ videa v 8:00", -500),
            ("Vyd√°n√≠ videa v 16:00", 1500),
            ("Vyd√°n√≠ videa o v√≠kendu", 800),
            ("N√°hodn√Ω ƒças vyd√°n√≠", 0)
        ]}
    ]
}

# Vytvo≈ôen√≠ hlavn√≠ho widget kontejneru
main_widget = widgets.VBox()

# Vytvo≈ôen√≠ nadpisu a instrukc√≠
nadpis = widgets.HTML("<h2>Hra: Z√≠skejte co nejv√≠ce div√°k≈Ø pro sv√© video!</h2>")
instrukce = widgets.HTML("<p>Vyberte nejlep≈°√≠ strategii pro maximalizaci sledovanosti va≈°eho videa. Ka≈æd√° volba m√° vliv na poƒçet va≈°ich div√°k≈Ø. Maxim√°ln√≠ potenci√°l je 8400 div√°k≈Ø.</p>")

# Vytvo≈ôen√≠ slovn√≠k≈Ø pro ukl√°d√°n√≠ widget≈Ø a jejich hodnot
widgets_dict = {}
radio_groups = {}

# Vytvo≈ôen√≠ widget≈Ø
accordion = widgets.Accordion()
accordion_children = []

for category, options in categories.items():
    category_box = widgets.VBox()
    category_items = []

    for option in options:
        if option["type"] == "checkbox":
            name = option["name"]
            value = option["value"]

            # Vytvo≈ôen√≠ checkbox widgetu
            cb = widgets.Checkbox(
                value=False,
                description=name,
                indent=False,
                layout=widgets.Layout(width="600px"),
                style={'description_width': 'initial'}
            )

            category_items.append(cb)

            # Ulo≈æen√≠ reference na widget
            widgets_dict[name] = {"widget": cb, "value": value, "type": "checkbox"}

        elif option["type"] == "radio":
            name = option["name"]
            radio_options = option["options"]

            # Vytvo≈ôen√≠ label pro skupinu
            group_label = widgets.HTML(f"<b>{name}:</b>")
            category_items.append(group_label)

            # P≈ôeveden√≠ mo≈ænost√≠ do form√°tu pro RadioButtons
            radio_labels = [label for label, _ in radio_options]
            radio_values = {label: value for label, value in radio_options}

            # Vytvo≈ôen√≠ RadioButtons widgetu
            rb = widgets.RadioButtons(
                options=radio_labels,
                layout=widgets.Layout(width="600px"),
                style={'description_width': 'initial'}
            )

            category_items.append(rb)

            # Ulo≈æen√≠ reference na widget
            radio_groups[name] = {"widget": rb, "values": radio_values, "type": "radio"}

    category_box.children = category_items
    accordion_children.append(category_box)

accordion.children = accordion_children
for i, category in enumerate(categories.keys()):
    accordion.set_title(i, category)

# Vytvo≈ôen√≠ progress baru
max_viewers = 8400
progress = widgets.IntProgress(
    value=0,
    min=0,
    max=max_viewers,
    description='Div√°ci:',
    bar_style='info',
    orientation='horizontal',
    layout=widgets.Layout(width='70%', height='30px')
)

viewer_count = widgets.Label('0 / 8400')
progress_box = widgets.HBox([progress, viewer_count])

# V√Ωstup pro graf a zpƒõtnou vazbu
graph_output = widgets.Output()
feedback_output = widgets.HTML("")

# Funkce pro v√Ωpoƒçet poƒçtu div√°k≈Ø
def calculate_viewers():
    total = 0
    selected_options = []

    # Zpracov√°n√≠ checkbox≈Ø
    for name, info in widgets_dict.items():
        if info["type"] == "checkbox" and info["widget"].value:
            total += info["value"]
            selected_options.append((name, info["value"]))

    # Zpracov√°n√≠ radio button≈Ø
    for name, info in radio_groups.items():
        if info["widget"].value is not None:
            selected_value = info["values"].get(info["widget"].value, 0)
            total += selected_value
            selected_options.append((info["widget"].value, selected_value))

    # Zaji≈°tƒõn√≠, ≈æe celkov√Ω poƒçet div√°k≈Ø nen√≠ z√°porn√Ω
    total = max(0, total)

    # Aktualizace progress baru
    progress.value = min(total, max_viewers)
    viewer_count.value = f"{total} / {max_viewers}"

    # Vytvo≈ôen√≠ grafu
    with graph_output:
        clear_output(wait=True)
        if selected_options:
            fig, ax = plt.subplots(figsize=(10, 6))

            options = [x[0] for x in selected_options]
            values = [x[1] for x in selected_options]
            colors = ['green' if v > 0 else 'red' for v in values]

            y_pos = range(len(options))
            ax.barh(y_pos, values, color=colors)
            ax.set_yticks(y_pos)
            ax.set_yticklabels(options)
            ax.invert_yaxis()  # Nejlep≈°√≠ polo≈æky naho≈ôe
            ax.set_title('Vliv jednotliv√Ωch voleb na poƒçet div√°k≈Ø')
            ax.axvline(x=0, color='black', linestyle='-', alpha=0.3)

            # P≈ôid√°n√≠ popisk≈Ø
            for i, v in enumerate(values):
                ax.text(v + (50 if v >= 0 else -50), i, str(v),
                        color='black', va='center', ha='left' if v >= 0 else 'right')

            plt.tight_layout()

            # P≈ôevod grafu na obr√°zek
            buf = io.BytesIO()
            plt.savefig(buf, format='png')
            buf.seek(0)
            img_str = base64.b64encode(buf.read()).decode('utf-8')
            plt.close(fig)

            display(HTML(f'<img src="data:image/png;base64,{img_str}" />'))

    # Poskytnut√≠ zpƒõtn√© vazby na z√°kladƒõ poƒçtu div√°k≈Ø
    feedback = ""
    if total <= 0:
        feedback = "‚ùå Va≈°e video je propad√°k. Zkuste jinou strategii!"
    elif total < 2000:
        feedback = "‚ö†Ô∏è Va≈°e video m√° jen velmi m√°lo div√°k≈Ø. Co zlep≈°it?"
    elif total < 4000:
        feedback = "üëç Dobr√© video, ale st√°le existuje prostor pro zlep≈°en√≠."
    elif total < 6000:
        feedback = "‚≠ê Va≈°e video je popul√°rn√≠! Jste na dobr√© cestƒõ."
    elif total < 8000:
        feedback = "üåü Skvƒõl√© video, kter√© m√° t√©mƒõ≈ô vir√°ln√≠ potenci√°l!"
    else:
        feedback = "üèÜ Gratulujeme! Va≈°e video je vir√°ln√≠ hit!"

    return total, feedback

# Tlaƒç√≠tka
calc_button = widgets.Button(
    description="Spoƒç√≠tej div√°ky",
    button_style="success",
    layout=widgets.Layout(width='200px')
)

reset_button = widgets.Button(
    description="Reset",
    button_style="warning",
    layout=widgets.Layout(width='100px')
)

button_box = widgets.HBox([calc_button, reset_button])

# Funkce pro tlaƒç√≠tka
def on_calc_button_click(b):
    total, feedback = calculate_viewers()
    feedback_output.value = f"<h3>{feedback}</h3>"

def on_reset_button_click(b):
    # Reset checkbox≈Ø
    for name, info in widgets_dict.items():
        if info["type"] == "checkbox":
            info["widget"].value = False

    # Reset radio button≈Ø
    for name, info in radio_groups.items():
        info["widget"].value = None

    # Reset v√Ωstupu
    progress.value = 0
    viewer_count.value = "0 / 8400"
    feedback_output.value = ""
    with graph_output:
        clear_output()

calc_button.on_click(on_calc_button_click)
reset_button.on_click(on_reset_button_click)

# Sestaven√≠ fin√°ln√≠ho UI - v≈°echny prvky vlo≈æ√≠me do hlavn√≠ho kontejneru
main_widget.children = [
    nadpis,
    instrukce,
    accordion,
    button_box,
    progress_box,
    feedback_output,
    graph_output
]

# Pouze jednou zobraz√≠me hlavn√≠ widget
display(main_widget)
```
