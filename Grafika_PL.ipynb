```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import matplotlib.pyplot as plt
import io
import base64
from IPython.display import Image

# Definice kategorií a možností s jejich bodovým ohodnocením
categories = {
    "Příprava videa": [
        {"type": "radio", "name": "Délka úvodu", "options": [
            ("Krátký úvod", 500),
            ("Dlouhý úvod", -1000),
            ("Bez úvodu", 0)
        ]},
        {"type": "checkbox", "name": "Točím se scénářem", "value": 300}
    ],
    "Kvalita produkce": [
        {"type": "radio", "name": "Technická kvalita", "options": [
            ("Kvalitní obraz, ale nekvalitní zvuk", -1500),
            ("Kvalitní zvuk, ale horší obraz", 1000),
            ("Natočím video se zvukem dohromady na jednom zařízení", -1500)
        ]}
    ],
    "Propagace": [
        {"type": "checkbox", "name": "Sdílení na sociálních sítích", "value": 2000},
        {"type": "radio", "name": "Náhledový obrázek", "options": [
            ("Udělám na YouTube pěknou náhledovku", 1500),
            ("Náhledovku na YouTube dělat nebudu", -500),
            ("Použiju generickou náhledovku", 100)
        ]},
        {"type": "checkbox", "name": "Dám video do playlistu", "value": 300}
    ],
    "Uživatelský zážitek": [
        {"type": "checkbox", "name": "Přidám titulky", "value": 500},
        {"type": "checkbox", "name": "Použiji vhodnou hudbu v pozadí", "value": 800},
        {"type": "checkbox", "name": "Příliš mnoho reklam", "value": -2000}
    ],
    "Načasování": [
        {"type": "radio", "name": "Čas vydání", "options": [
            ("Vydání videa v 8:00", -500),
            ("Vydání videa v 16:00", 1500),
            ("Vydání videa o víkendu", 800),
            ("Náhodný čas vydání", 0)
        ]}
    ]
}

# Vytvoření hlavního widget kontejneru
main_widget = widgets.VBox()

# Vytvoření nadpisu a instrukcí
nadpis = widgets.HTML("<h2>Hra: Získejte co nejvíce diváků pro své video!</h2>")
instrukce = widgets.HTML("<p>Vyberte nejlepší strategii pro maximalizaci sledovanosti vašeho videa. Každá volba má vliv na počet vašich diváků. Maximální potenciál je 8400 diváků.</p>")

# Vytvoření slovníků pro ukládání widgetů a jejich hodnot
widgets_dict = {}
radio_groups = {}

# Vytvoření widgetů
accordion = widgets.Accordion()
accordion_children = []

for category, options in categories.items():
    category_box = widgets.VBox()
    category_items = []

    for option in options:
        if option["type"] == "checkbox":
            name = option["name"]
            value = option["value"]

            # Vytvoření checkbox widgetu
            cb = widgets.Checkbox(
                value=False,
                description=name,
                indent=False,
                layout=widgets.Layout(width="600px"),
                style={'description_width': 'initial'}
            )

            category_items.append(cb)

            # Uložení reference na widget
            widgets_dict[name] = {"widget": cb, "value": value, "type": "checkbox"}

        elif option["type"] == "radio":
            name = option["name"]
            radio_options = option["options"]

            # Vytvoření label pro skupinu
            group_label = widgets.HTML(f"<b>{name}:</b>")
            category_items.append(group_label)

            # Převedení možností do formátu pro RadioButtons
            radio_labels = [label for label, _ in radio_options]
            radio_values = {label: value for label, value in radio_options}

            # Vytvoření RadioButtons widgetu
            rb = widgets.RadioButtons(
                options=radio_labels,
                layout=widgets.Layout(width="600px"),
                style={'description_width': 'initial'}
            )

            category_items.append(rb)

            # Uložení reference na widget
            radio_groups[name] = {"widget": rb, "values": radio_values, "type": "radio"}

    category_box.children = category_items
    accordion_children.append(category_box)

accordion.children = accordion_children
for i, category in enumerate(categories.keys()):
    accordion.set_title(i, category)

# Vytvoření progress baru
max_viewers = 8400
progress = widgets.IntProgress(
    value=0,
    min=0,
    max=max_viewers,
    description='Diváci:',
    bar_style='info',
    orientation='horizontal',
    layout=widgets.Layout(width='70%', height='30px')
)

viewer_count = widgets.Label('0 / 8400')
progress_box = widgets.HBox([progress, viewer_count])

# Výstup pro graf a zpětnou vazbu
graph_output = widgets.Output()
feedback_output = widgets.HTML("")

# Funkce pro výpočet počtu diváků
def calculate_viewers():
    total = 0
    selected_options = []

    # Zpracování checkboxů
    for name, info in widgets_dict.items():
        if info["type"] == "checkbox" and info["widget"].value:
            total += info["value"]
            selected_options.append((name, info["value"]))

    # Zpracování radio buttonů
    for name, info in radio_groups.items():
        if info["widget"].value is not None:
            selected_value = info["values"].get(info["widget"].value, 0)
            total += selected_value
            selected_options.append((info["widget"].value, selected_value))

    # Zajištění, že celkový počet diváků není záporný
    total = max(0, total)

    # Aktualizace progress baru
    progress.value = min(total, max_viewers)
    viewer_count.value = f"{total} / {max_viewers}"

    # Vytvoření grafu
    with graph_output:
        clear_output(wait=True)
        if selected_options:
            fig, ax = plt.subplots(figsize=(10, 6))

            options = [x[0] for x in selected_options]
            values = [x[1] for x in selected_options]
            colors = ['green' if v > 0 else 'red' for v in values]

            y_pos = range(len(options))
            ax.barh(y_pos, values, color=colors)
            ax.set_yticks(y_pos)
            ax.set_yticklabels(options)
            ax.invert_yaxis()  # Nejlepší položky nahoře
            ax.set_title('Vliv jednotlivých voleb na počet diváků')
            ax.axvline(x=0, color='black', linestyle='-', alpha=0.3)

            # Přidání popisků
            for i, v in enumerate(values):
                ax.text(v + (50 if v >= 0 else -50), i, str(v),
                        color='black', va='center', ha='left' if v >= 0 else 'right')

            plt.tight_layout()

            # Převod grafu na obrázek
            buf = io.BytesIO()
            plt.savefig(buf, format='png')
            buf.seek(0)
            img_str = base64.b64encode(buf.read()).decode('utf-8')
            plt.close(fig)

            display(HTML(f'<img src="data:image/png;base64,{img_str}" />'))

    # Poskytnutí zpětné vazby na základě počtu diváků
    feedback = ""
    if total <= 0:
        feedback = "❌ Vaše video je propadák. Zkuste jinou strategii!"
    elif total < 2000:
        feedback = "⚠️ Vaše video má jen velmi málo diváků. Co zlepšit?"
    elif total < 4000:
        feedback = "👍 Dobré video, ale stále existuje prostor pro zlepšení."
    elif total < 6000:
        feedback = "⭐ Vaše video je populární! Jste na dobré cestě."
    elif total < 8000:
        feedback = "🌟 Skvělé video, které má téměř virální potenciál!"
    else:
        feedback = "🏆 Gratulujeme! Vaše video je virální hit!"

    return total, feedback

# Tlačítka
calc_button = widgets.Button(
    description="Spočítej diváky",
    button_style="success",
    layout=widgets.Layout(width='200px')
)

reset_button = widgets.Button(
    description="Reset",
    button_style="warning",
    layout=widgets.Layout(width='100px')
)

button_box = widgets.HBox([calc_button, reset_button])

# Funkce pro tlačítka
def on_calc_button_click(b):
    total, feedback = calculate_viewers()
    feedback_output.value = f"<h3>{feedback}</h3>"

def on_reset_button_click(b):
    # Reset checkboxů
    for name, info in widgets_dict.items():
        if info["type"] == "checkbox":
            info["widget"].value = False

    # Reset radio buttonů
    for name, info in radio_groups.items():
        info["widget"].value = None

    # Reset výstupu
    progress.value = 0
    viewer_count.value = "0 / 8400"
    feedback_output.value = ""
    with graph_output:
        clear_output()

calc_button.on_click(on_calc_button_click)
reset_button.on_click(on_reset_button_click)

# Sestavení finálního UI - všechny prvky vložíme do hlavního kontejneru
main_widget.children = [
    nadpis,
    instrukce,
    accordion,
    button_box,
    progress_box,
    feedback_output,
    graph_output
]

# Pouze jednou zobrazíme hlavní widget
display(main_widget)
```

```python
# @title
# Funkce pro načtení obrázku z URL
def nacti_obrazek_z_url(url):
    response = requests.get(url)
    return BytesIO(response.content).getvalue()

# URL adresy obrázků
objektiv_url = "https://1313650871.rsc.cdn77.org/getthumbnail.aspx?q=60&height=100000&width=100000&id_file=894238101"
clona_url = "https://lh5.googleusercontent.com/proxy/qvCl3XVFUdAqvcFyhCxh2GUpFfhEw78Z3Xv658gyAOqF42RtBrnpxjgCMKofTBVvkEexdIKQtlOsK2dmEJ6FY23qeWet4_GNHVT0cJFgJLaV_w"
senzor_url = "https://cdn.myshoptet.com/usr/www.phototools.cz/user/documents/upload/slovnik/cmos-senzor.png"
telo_url = "https://www.vkfoto.cz/imgcache/b/b/pvba560ae-d780-telo-1_540_540_108460.jpg"

# Seznam komponent pro dropdown - zamícháme je
komponenty = ['Objektiv', 'Clona', 'Senzor', 'Tělo fotoaparátu']
# Zamíchání možností
random.shuffle(komponenty)

# Přidáme prázdnou možnost jako první volbu
komponenty_s_prazdnou = ['-- Vyberte --'] + komponenty

# Správné odpovědi pro kontrolu
spravne_odpovedi = {
    'img1': 'Objektiv',
    'img2': 'Clona',
    'img3': 'Senzor',
    'img4': 'Tělo fotoaparátu'
}

# Vytvoření widgetů pro obrázky
try:
    img1 = widgets.Image(
        value=nacti_obrazek_z_url(objektiv_url),
        format='jpg',
        width=250,
        height=250,
    )

    img2 = widgets.Image(
        value=nacti_obrazek_z_url(clona_url),
        format='jpg',
        width=250,
        height=250,
    )

    img3 = widgets.Image(
        value=nacti_obrazek_z_url(senzor_url),
        format='png',
        width=250,
        height=250,
    )

    img4 = widgets.Image(
        value=nacti_obrazek_z_url(telo_url),
        format='jpg',
        width=250,
        height=250,
    )

    # Vytvoření dropdown menu pro každý obrázek s prázdnou výchozí hodnotou
    dropdown1 = widgets.Dropdown(options=komponenty_s_prazdnou, value='-- Vyberte --', description='Komponenta:')
    dropdown2 = widgets.Dropdown(options=komponenty_s_prazdnou, value='-- Vyberte --', description='Komponenta:')
    dropdown3 = widgets.Dropdown(options=komponenty_s_prazdnou, value='-- Vyberte --', description='Komponenta:')
    dropdown4 = widgets.Dropdown(options=komponenty_s_prazdnou, value='-- Vyberte --', description='Komponenta:')

    # Výsledkové labely
    result1 = widgets.Label('')
    result2 = widgets.Label('')
    result3 = widgets.Label('')
    result4 = widgets.Label('')

    # Funkce pro kontrolu odpovědí
    def kontrola_odpovedi(btn):
        spravne = 0

        # Kontrola, zda jsou všechna pole vyplněna
        if dropdown1.value == '-- Vyberte --' or dropdown2.value == '-- Vyberte --' or dropdown3.value == '-- Vyberte --' or dropdown4.value == '-- Vyberte --':
            celkovy_vysledek.value = "Prosím vyplňte všechna pole!"
            return

        if dropdown1.value == spravne_odpovedi['img1']:
            result1.value = '✓ Správně!'
            spravne += 1
        else:
            result1.value = '✗ Nesprávně. Toto je ' + spravne_odpovedi['img1']

        if dropdown2.value == spravne_odpovedi['img2']:
            result2.value = '✓ Správně!'
            spravne += 1
        else:
            result2.value = '✗ Nesprávně. Toto je ' + spravne_odpovedi['img2']

        if dropdown3.value == spravne_odpovedi['img3']:
            result3.value = '✓ Správně!'
            spravne += 1
        else:
            result3.value = '✗ Nesprávně. Toto je ' + spravne_odpovedi['img3']

        if dropdown4.value == spravne_odpovedi['img4']:
            result4.value = '✓ Správně!'
            spravne += 1
        else:
            result4.value = '✗ Nesprávně. Toto je ' + spravne_odpovedi['img4']

        celkovy_vysledek.value = f"Celkový výsledek: {spravne} správně z {len(spravne_odpovedi)}"

    # Tlačítko pro kontrolu
    kontrola_btn = widgets.Button(
        description='Kontrola odpovědí',
        button_style='primary',
        icon='check'
    )
    kontrola_btn.on_click(kontrola_odpovedi)

    # Tlačítko pro reset
    def reset_odpovedi(btn):
        dropdown1.value = '-- Vyberte --'
        dropdown2.value = '-- Vyberte --'
        dropdown3.value = '-- Vyberte --'
        dropdown4.value = '-- Vyberte --'
        result1.value = ''
        result2.value = ''
        result3.value = ''
        result4.value = ''
        celkovy_vysledek.value = ''

    reset_btn = widgets.Button(
        description='Reset',
        button_style='warning',
        icon='refresh'
    )
    reset_btn.on_click(reset_odpovedi)

    # Celkový výsledek
    celkovy_vysledek = widgets.Label('')

    # Vytvoření boxů pro každý obrázek s dropdown
    box1 = widgets.VBox([img1, dropdown1, result1])
    box2 = widgets.VBox([img2, dropdown2, result2])
    box3 = widgets.VBox([img3, dropdown3, result3])
    box4 = widgets.VBox([img4, dropdown4, result4])

    # Uspořádání všech komponent do gridu 2x2
    horni_rada = widgets.HBox([box1, box2])
    dolni_rada = widgets.HBox([box3, box4])
    fotoaparat_komponenty = widgets.VBox([horni_rada, dolni_rada])

    # Tlačítka vedle sebe
    tlacitka = widgets.HBox([kontrola_btn, reset_btn])

    # Přidání nadpisu
    nadpis = widgets.HTML(value="<h2>Určete komponenty fotoaparátu</h2>")
    instrukce = widgets.HTML(value="<p>Vyberte správnou komponentu ke každému obrázku a klikněte na tlačítko 'Kontrola odpovědí'.</p>")

    # Složení finálního widgetu
    finalni_widget = widgets.VBox([nadpis, instrukce, fotoaparat_komponenty, tlacitka, celkovy_vysledek])

    # Zobrazení výsledného widgetu
    display(finalni_widget)

except Exception as e:
    print(f"Došlo k chybě při načítání obrázků: {e}")
    print("Zkontrolujte, zda jsou URL adresy správné a zda máte nainstalovaný modul requests.")
```
