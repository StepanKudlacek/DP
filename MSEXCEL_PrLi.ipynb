```python
# @title
# Import potřebných modulů
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import random

# Na konci kódu přidáme výstup, že vše je načteno:
display(widgets.HTML("<h3 style='color: green;'>Pokračuj, vše je staženo.</h3>"))
```

```python
# @title

# ===============================
# 1. Úkol: Základní formátování buněk
# ===============================

# --- Část 1: Výběr z rolety pro změnu barvy písma ---
options_q1 = ["Výplň", "Zarovnání", "Barva písma"]
random.shuffle(options_q1)  # Zamícháme pořadí možností

dropdown_q1 = widgets.Dropdown(
    options=options_q1,
    value=None,
    description='',  # Popisek zobrazíme zvlášť
    layout=widgets.Layout(width='150px')
)

button_q1 = widgets.Button(
    description="Ověřit odpověď",
    button_style='info'
)

output_q1 = widgets.Output()

def check_dropdown(b):
    with output_q1:
        clear_output()
        if dropdown_q1.value == "Barva písma":
            print("Správně!")
        else:
            print("Nesprávně, zkuste to znovu.")

button_q1.on_click(check_dropdown)

display(widgets.HTML("<h3>1. Základní formátování buněk</h3>"))
display(widgets.HTML("<b>Která volba se používá pro změnu barvy písma?</b>"))
display(dropdown_q1, button_q1, output_q1)

# --- Část 2: Doplňovačka s textovým vstupem ---
# Otázka: "Pro nastavení tučného písma se používá ikona ______."
text_input_q2 = widgets.Text(
    value="",
    placeholder="Napište zde (odpovědí je jedno písmeno)",
    description='',  # Nepoužíváme description, abychom mohli text zobrazit zvlášť
)

button_q2 = widgets.Button(
    description="Ověřit odpověď",
    button_style='info'
)

output_q2 = widgets.Output()

expected_answer = "B"

def check_text_input(b):
    with output_q2:
        clear_output()
        if text_input_q2.value.strip().upper() == expected_answer:
            print("Správně!")
        else:
            print("Nesprávně, zkuste to znovu.")

button_q2.on_click(check_text_input)

display(widgets.HTML("<br><b>Pro nastavení tučného písma se používá ikona ______.</b>"))
display(text_input_q2, button_q2, output_q2)

# ===============================
# 3. Úkol: Přiřaďte pojmy k jejich funkcím
# ===============================

display(widgets.HTML("<br><b>Přiřaďte pojmy k jejich funkcím:</b>"))

# Definujeme vlastní sady možností pro každý dropdown a správné odpovědi.
# 1) Okraje
okraje_options = [
    "Nastavení barvy písma",
    "Nastavení ohraničení buněk",
    "Zarovnání na střed",
    "Nastavení barvy pozadí"
]
okraje_correct = "Nastavení ohraničení buněk"

# 2) Výplň
vypln_options = [
    "Nastavení barvy pozadí",
    "Ctrl+S",
    "Nastavení sloupců",
    "Nastavení ohraničení buněk"
]
vypln_correct = "Nastavení barvy pozadí"

# 3) Klávesová zkratka pro uložení souboru
zkratka_options = [
    "Ctrl+S",
    "Ctrl+Z",
    "Ctrl+B",
    "Ctrl+Shift+F"
]
zkratka_correct = "Ctrl+S"

def create_dropdown(options):
    # Zduplikujeme seznam, aby shuffle neměnil původní
    opts = options.copy()
    random.shuffle(opts)
    dd = widgets.Dropdown(
        options=opts,
        value=None,
        description='',  # Popisek přidáme zvlášť jako HTML
        layout=widgets.Layout(width='150px')
    )
    return dd

# Vytvoříme dropdowny
dropdown_okraje = create_dropdown(okraje_options)
dropdown_vypln = create_dropdown(vypln_options)
dropdown_zkratka = create_dropdown(zkratka_options)

# Popisky k dropdownům
okraje_label = widgets.HTML("<b>Okraje</b>")
okraje_box = widgets.VBox([okraje_label, dropdown_okraje])

vypln_label = widgets.HTML("<b>Výplň</b>")
vypln_box = widgets.VBox([vypln_label, dropdown_vypln])

zkratka_label = widgets.HTML("<b>Klávesová zkratka pro uložení souboru</b>")
zkratka_box = widgets.VBox([zkratka_label, dropdown_zkratka])

button_q3 = widgets.Button(
    description="Ověřit odpovědi",
    button_style='info'
)

output_q3 = widgets.Output()

def check_assignment(b):
    with output_q3:
        clear_output()
        correct = True

        # Ověření "Okraje"
        if dropdown_okraje.value != okraje_correct:
            print("Okraje: Nesprávně")
            correct = False
        else:
            print("Okraje: Správně")

        # Ověření "Výplň"
        if dropdown_vypln.value != vypln_correct:
            print("Výplň: Nesprávně")
            correct = False
        else:
            print("Výplň: Správně")

        # Ověření "Klávesová zkratka"
        if dropdown_zkratka.value != zkratka_correct:
            print("Klávesová zkratka: Nesprávně")
            correct = False
        else:
            print("Klávesová zkratka: Správně")

        if correct:
            print("Všechny odpovědi jsou správné!")
        else:
            print("Zkuste to prosím znovu.")

button_q3.on_click(check_assignment)

# Uspořádání boxů do mřížky
assignment_box = widgets.GridBox(
    children=[okraje_box, vypln_box, zkratka_box],
    layout=widgets.Layout(
        grid_template_columns="repeat(3, 33%)",
        grid_gap="10px 10px"
    )
)

display(assignment_box)
display(button_q3, output_q3)
```

```python
# @title
########################################
# 3. Podmíněné formátování
########################################

display(widgets.HTML("<h2>2. Podmíněné formátování</h2>"))

# --- Otázka 1: RadioButtons ---
display(widgets.HTML("<b>Otázka 1:</b> Kterou záložku v nabídce podmíněného formátování použijete, abyste mohli nastavit pravidla pro zvýraznění buněk, které obsahují text AHOJ "))

q3_1_radio = widgets.RadioButtons(
    options=[
        "Text, který obsahuje",
        "Rovná se",
        "Použít sadu ikon"
    ],
    description='',
    layout=widgets.Layout(width='auto')
)

q3_1_button = widgets.Button(description="Ověřit odpověď", button_style='info')
q3_1_output = widgets.Output()

def check_q3_1(b):
    with q3_1_output:
        clear_output()
        # Správná odpověď je "Rovná se"
        if q3_1_radio.value == "Rovná se":
            print("Správně! Abyste mohli najít buňky, které obsahují pouze konkrétní text, museli byste použít nabídku ROVNÁ SE.“")
        else:
            print("Nesprávně, zkuste to znovu.")

q3_1_button.on_click(check_q3_1)

display(q3_1_radio, q3_1_button, q3_1_output)

# --- Otázka 2: TextInput ---
display(widgets.HTML("<br><b>Otázka 2:</b> Jaké podmíněné formátování byste zvolili, chcete-li zobrazit plynulý přechod barev podle velikosti hodnot? (Napište název funkce/pravidla)"))

q3_2_text = widgets.Text(
    value="",
    placeholder="Anglicky se to řekne Color Scale",
    description=''
)
q3_2_button = widgets.Button(description="Ověřit odpověď", button_style='info')
q3_2_output = widgets.Output()

def check_q3_2(b):
    with q3_2_output:
        clear_output()
        # Očekávaná odpověď je "Barevná škála"
        if q3_2_text.value.strip().lower() in ["barevná škála", "barevna skala", "color scale"]:
            print("Správně! Pro plynulý přechod se využívá barevná škála.")
        else:
            print("Nesprávně, zkuste to znovu.")

q3_2_button.on_click(check_q3_2)

display(q3_2_text, q3_2_button, q3_2_output)

# --- Otázka 3: Přiřazovací cvičení (GridBox) ---
display(widgets.HTML("<br><b>Otázka 3:</b> Přiřaďte název podmíněného formátování k popisu."))

# Tři pojmy + tři popisy
# 1) Ikonové sady -> "Zobrazují různé ikony podle rozsahu hodnot"
# 2) Zvýraznění pravidel -> "Podle konkrétních podmínek (větší/menší než...)"
# 3) Barevné škály -> "Plynulý přechod barev podle velikosti hodnot"

ikonove_sady_label = widgets.HTML("<b>Ikonové sady</b>")
zvyrazneni_label = widgets.HTML("<b>Zvýraznění pravidel</b>")
barevne_skaly_label = widgets.HTML("<b>Barevné škály</b>")

# Každý dropdown má jiné možnosti
ikonove_sady_options = [
    "Plynulý přechod barev podle velikosti hodnot",
    "Zobrazují různé ikony podle rozsahu hodnot",
    "Podle konkrétních podmínek (větší/menší než...)"
]
zvyrazneni_options = ikonove_sady_options.copy()
barevne_skaly_options = ikonove_sady_options.copy()

random.shuffle(ikonove_sady_options)
random.shuffle(zvyrazneni_options)
random.shuffle(barevne_skaly_options)

ikonove_sady_dd = widgets.Dropdown(options=ikonove_sady_options, value=None, layout=widgets.Layout(width='220px'))
zvyrazneni_dd = widgets.Dropdown(options=zvyrazneni_options, value=None, layout=widgets.Layout(width='220px'))
barevne_skaly_dd = widgets.Dropdown(options=barevne_skaly_options, value=None, layout=widgets.Layout(width='220px'))

ikonove_sady_box = widgets.VBox([ikonove_sady_label, ikonove_sady_dd])
zvyrazneni_box = widgets.VBox([zvyrazneni_label, zvyrazneni_dd])
barevne_skaly_box = widgets.VBox([barevne_skaly_label, barevne_skaly_dd])

q3_3_button = widgets.Button(description="Ověřit odpověď", button_style='info')
q3_3_output = widgets.Output()

def check_q3_3(b):
    with q3_3_output:
        clear_output()
        correct = True

        # Ikonové sady
        if ikonove_sady_dd.value != "Zobrazují různé ikony podle rozsahu hodnot":
            print("Ikonové sady: Nesprávně")
            correct = False
        else:
            print("Ikonové sady: Správně")

        # Zvýraznění pravidel
        if zvyrazneni_dd.value != "Podle konkrétních podmínek (větší/menší než...)":
            print("Zvýraznění pravidel: Nesprávně")
            correct = False
        else:
            print("Zvýraznění pravidel: Správně")

        # Barevné škály
        if barevne_skaly_dd.value != "Plynulý přechod barev podle velikosti hodnot":
            print("Barevné škály: Nesprávně")
            correct = False
        else:
            print("Barevné škály: Správně")

        if correct:
            print("Všechny odpovědi jsou správné!")
        else:
            print("Zkuste to prosím znovu.")

q3_3_button.on_click(check_q3_3)

q3_3_grid = widgets.GridBox(
    children=[ikonove_sady_box, zvyrazneni_box, barevne_skaly_box],
    layout=widgets.Layout(
        grid_template_columns="repeat(3, 1fr)",
        grid_gap="10px"
    )
)

display(q3_3_grid)
display(q3_3_button, q3_3_output)
```

```python
# @title
########################################
# 4. Základní vzorce a matematické operace
########################################

display(widgets.HTML("<h2>4. Základní vzorce a matematické operace</h2>"))

# --- Otázka 1: TextInput ---
display(widgets.HTML("<b>Otázka 1:</b> Jak vypadá vzorec pro sečtení hodnot v buňkách A1 a B1?"))

q4_1_text = widgets.Text(
    value="",
    description=''
)
q4_1_button = widgets.Button(description="Ověřit odpověď", button_style='info')
q4_1_output = widgets.Output()

def check_q4_1(b):
    with q4_1_output:
        clear_output()
        # Očekávaná odpověď
        correct_answers = ["=a1+b1"]
        if q4_1_text.value.strip().lower().replace(" ", "") in correct_answers:
            print("Správně! Základní sčítání se zadává jako =A1+B1.")
        else:
            print("Nesprávně, zkuste to znovu.")

q4_1_button.on_click(check_q4_1)

display(q4_1_text, q4_1_button, q4_1_output)

# --- Otázka 2: RadioButtons (absolutní odkaz) ---
display(widgets.HTML("<br><b>Otázka 2:</b> Který zápis představuje absolutní odkaz na buňku C3? Absolutní odkaz zafixuje jak řádek, tak sloupec."))

q4_2_radio = widgets.RadioButtons(
    options=["C3", "$C$3", "C$3", "$C3"],
    description='',
    layout=widgets.Layout(width='auto')
)

q4_2_button = widgets.Button(description="Ověřit odpověď", button_style='info')
q4_2_output = widgets.Output()

def check_q4_2(b):
    with q4_2_output:
        clear_output()
        if q4_2_radio.value == "$C$3":
            print("Správně! Absolutní odkaz se zapisuje jako $C$3.")
        else:
            print("Nesprávně, zkuste to znovu.")

q4_2_button.on_click(check_q4_2)

display(q4_2_radio, q4_2_button, q4_2_output)

# --- Otázka 3: Přiřaďte názvy operací k příslušným symbolům (opravená verze) ---
display(widgets.HTML("<br><b>Otázka 3:</b> Přiřaďte správný symbol k názvu operace."))

# Popisky pro jednotlivé operace
scitani_label = widgets.HTML("<b>Sčítání</b>")
odcitani_label = widgets.HTML("<b>Odčítání</b>")
nasobeni_label = widgets.HTML("<b>Násobení</b>")
deleni_label = widgets.HTML("<b>Dělení</b>")

# Definujeme pro každou operaci vlastní seznam možností
scitani_options = ["*", "/", "–", "+"]
odcitani_options = ["/", "+", "-", "÷"]
nasobeni_options = ["*", "x", "×"]
deleni_options = ["/", "÷", ":"]

def create_operator_dropdown(options):
    opts = options.copy()
    random.shuffle(opts)
    return widgets.Dropdown(options=opts, value=None, layout=widgets.Layout(width='80px'))

# Vytvoříme dropdowny s příslušnými možnostmi
scitani_dd = create_operator_dropdown(scitani_options)
odcitani_dd = create_operator_dropdown(odcitani_options)
nasobeni_dd = create_operator_dropdown(nasobeni_options)
deleni_dd = create_operator_dropdown(deleni_options)

# Uspořádáme každou operaci do VBox (popisek nad dropdownem)
scitani_box = widgets.VBox([scitani_label, scitani_dd])
odcitani_box = widgets.VBox([odcitani_label, odcitani_dd])
nasobeni_box = widgets.VBox([nasobeni_label, nasobeni_dd])
deleni_box = widgets.VBox([deleni_label, deleni_dd])

# Tlačítko a výstup pro kontrolu odpovědí
q4_3_button = widgets.Button(description="Ověřit odpovědi", button_style='info')
q4_3_output = widgets.Output()

def check_q4_3(b):
    with q4_3_output:
        clear_output()
        correct = True

        if scitani_dd.value != "+":
            print("Sčítání: Nesprávně")
            correct = False
        else:
            print("Sčítání: Správně")

        if odcitani_dd.value != "-":
            print("Odčítání: Nesprávně")
            correct = False
        else:
            print("Odčítání: Správně")

        if nasobeni_dd.value != "*":
            print("Násobení: Nesprávně")
            correct = False
        else:
            print("Násobení: Správně")

        if deleni_dd.value != "/":
            print("Dělení: Nesprávně")
            correct = False
        else:
            print("Dělení: Správně")

        if correct:
            print("Všechny odpovědi jsou správné!")
        else:
            print("Zkuste to prosím znovu.")

q4_3_button.on_click(check_q4_3)

# Uspořádání do mřížky (4 sloupce)
q4_3_grid = widgets.GridBox(
    children=[scitani_box, odcitani_box, nasobeni_box, deleni_box],
    layout=widgets.Layout(
        grid_template_columns="repeat(4, 1fr)",
        grid_gap="10px"
    )
)

display(q4_3_grid)
display(q4_3_button, q4_3_output)
```

```python
# @title
########################################
# 5. Funkce
########################################

display(widgets.HTML("<h2>5. Funkce</h2>"))

# --- Otázka 1: RadioButtons ---
display(widgets.HTML("<b>Otázka 1:</b> Která funkce vrátí nejvyšší hodnotu v daném rozsahu?"))

q5_1_radio = widgets.RadioButtons(
    options=["MIN", "MAX", "PRŮMĚR", "SUMA"],
    description=''
)
q5_1_button = widgets.Button(description="Ověřit odpověď", button_style='info')
q5_1_output = widgets.Output()

def check_q5_1(b):
    with q5_1_output:
        clear_output()
        if q5_1_radio.value == "MAX":
            print("Správně! MAX vrací nejvyšší hodnotu v daném rozsahu.")
        else:
            print("Nesprávně, zkuste to znovu.")

q5_1_button.on_click(check_q5_1)

display(q5_1_radio, q5_1_button, q5_1_output)

# --- Otázka 2: TextInput ---
display(widgets.HTML("<br><b>Otázka 2:</b> Jak byste zapsali funkci pro součet hodnot v rozsahu A1:A5?"))

q5_2_text = widgets.Text(
    value="",
    description=''
)
q5_2_button = widgets.Button(description="Ověřit odpověď", button_style='info')
q5_2_output = widgets.Output()

def check_q5_2(b):
    with q5_2_output:
        clear_output()
        # Očekávaná odpověď (ignorujeme velikost písmen a mezery)
        correct_form = "=suma(a1:a5)"
        user_answer = q5_2_text.value.strip().lower().replace(" ", "")
        if user_answer == correct_form:
            print("Správně! Funkce pro součet je =SUMA(A1:A5).")
        else:
            print("Nesprávně, zkuste to znovu.")

q5_2_button.on_click(check_q5_2)

display(q5_2_text, q5_2_button, q5_2_output)

# --- Otázka 3: Přiřazovací cvičení (MIN, AVERAGE, COUNT) ---
display(widgets.HTML("<br><b>Otázka 3:</b> Přiřaďte funkci k jejímu popisu."))

# Pojmy: "MIN", "AVERAGE (PRŮMĚR)", "COUNT (POČET)"
# Popisy:
# 1) "Vrátí nejnižší hodnotu" -> MIN
# 2) "Vrátí průměr hodnot" -> AVERAGE
# 3) "Spočítá počet číselných hodnot" -> COUNT

min_label = widgets.HTML("<b>MIN</b>")
avg_label = widgets.HTML("<b>AVERAGE (PRŮMĚR)</b>")
count_label = widgets.HTML("<b>COUNT (POČET)</b>")

desc_options = [
    "Vrátí nejnižší hodnotu",
    "Vrátí průměr hodnot",
    "Spočítá počet číselných hodnot"
]

def create_desc_dropdown():
    opts = desc_options.copy()
    random.shuffle(opts)
    return widgets.Dropdown(options=opts, value=None, layout=widgets.Layout(width='200px'))

min_dd = create_desc_dropdown()
avg_dd = create_desc_dropdown()
count_dd = create_desc_dropdown()

min_box = widgets.VBox([min_label, min_dd])
avg_box = widgets.VBox([avg_label, avg_dd])
count_box = widgets.VBox([count_label, count_dd])

q5_3_button = widgets.Button(description="Ověřit odpovědi", button_style='info')
q5_3_output = widgets.Output()

def check_q5_3(b):
    with q5_3_output:
        clear_output()
        correct = True

        if min_dd.value != "Vrátí nejnižší hodnotu":
            print("MIN: Nesprávně")
            correct = False
        else:
            print("MIN: Správně")

        if avg_dd.value != "Vrátí průměr hodnot":
            print("AVERAGE: Nesprávně")
            correct = False
        else:
            print("AVERAGE: Správně")

        if count_dd.value != "Spočítá počet číselných hodnot":
            print("COUNT: Nesprávně")
            correct = False
        else:
            print("COUNT: Správně")

        if correct:
            print("Všechny odpovědi jsou správné!")
        else:
            print("Zkuste to prosím znovu.")

q5_3_button.on_click(check_q5_3)

q5_3_grid = widgets.GridBox(
    children=[min_box, avg_box, count_box],
    layout=widgets.Layout(
        grid_template_columns="repeat(3, 1fr)",
        grid_gap="10px"
    )
)

display(q5_3_grid)
display(q5_3_button, q5_3_output)
```

```python
# @title
########################################
# 7. Podmíněné součty a průměry
########################################

display(widgets.HTML("<h2>6. Podmíněné součty a průměry</h2>"))

# --- Otázka 1: RadioButtons ---
display(widgets.HTML("<b>Otázka 1:</b> Která funkce spočítá počet buněk splňujících zadané kritérium?"))

q7_1_radio = widgets.RadioButtons(
    options=["SUMIF", "COUNTIF", "AVERAGEIF"],
    description=''
)
q7_1_button = widgets.Button(description="Ověřit odpověď", button_style='info')
q7_1_output = widgets.Output()

def check_q7_1(b):
    with q7_1_output:
        clear_output()
        if q7_1_radio.value == "COUNTIF":
            print("Správně! COUNTIF počítá počet buněk splňujících kritérium.")
        else:
            print("Nesprávně, zkuste to znovu.")

q7_1_button.on_click(check_q7_1)

display(q7_1_radio, q7_1_button, q7_1_output)

# --- Otázka 2: TextInput ---
display(widgets.HTML("<br><b>Otázka 2:</b> Zapište vzorec pro SUMIF, který sečte hodnoty větší než 50 v rozsahu A1:A10."))

q7_2_text = widgets.Text(
    value="",
    description=''
)
q7_2_button = widgets.Button(description="Ověřit odpověď", button_style='info')
q7_2_output = widgets.Output()

def check_q7_2(b):
    with q7_2_output:
        clear_output()
        correct_answer = '=sumif(a1:a10;">50")'
        user_answer = q7_2_text.value.strip().lower().replace(" ", "")
        if user_answer == correct_answer:
            print("Správně! =SUMIF(A1:A10, \">50\") sečte všechny hodnoty v rozsahu A1:A10, které jsou větší než 50.")
        else:
            print("Nesprávně, zkuste to znovu.")

q7_2_button.on_click(check_q7_2)

display(q7_2_text, q7_2_button, q7_2_output)

# --- Otázka 3: Přiřazovací cvičení (SUMIF, COUNTIF, AVERAGEIF) ---
display(widgets.HTML("<br><b>Otázka 3:</b> Přiřaďte funkci k jejímu popisu."))

# 1) SUMIF -> "Sečte hodnoty splňující kritérium"
# 2) COUNTIF -> "Spočítá počet buněk splňujících kritérium"
# 3) AVERAGEIF -> "Vypočítá průměr hodnot na základě kritéria"

sumif_label = widgets.HTML("<b>SUMIF</b>")
countif_label = widgets.HTML("<b>COUNTIF</b>")
averageif_label = widgets.HTML("<b>AVERAGEIF</b>")

desc_7_3 = [
    "Sečte hodnoty splňující kritérium",
    "Spočítá počet buněk splňujících kritérium",
    "Vypočítá průměr hodnot na základě kritéria"
]

def create_7_3_dropdown():
    opts = desc_7_3.copy()
    random.shuffle(opts)
    return widgets.Dropdown(options=opts, value=None, layout=widgets.Layout(width='250px'))

sumif_dd = create_7_3_dropdown()
countif_dd = create_7_3_dropdown()
averageif_dd = create_7_3_dropdown()

sumif_box = widgets.VBox([sumif_label, sumif_dd])
countif_box = widgets.VBox([countif_label, countif_dd])
averageif_box = widgets.VBox([averageif_label, averageif_dd])

q7_3_button = widgets.Button(description="Ověřit odpovědi", button_style='info')
q7_3_output = widgets.Output()

def check_q7_3(b):
    with q7_3_output:
        clear_output()
        correct = True

        if sumif_dd.value != "Sečte hodnoty splňující kritérium":
            print("SUMIF: Nesprávně")
            correct = False
        else:
            print("SUMIF: Správně")

        if countif_dd.value != "Spočítá počet buněk splňujících kritérium":
            print("COUNTIF: Nesprávně")
            correct = False
        else:
            print("COUNTIF: Správně")

        if averageif_dd.value != "Vypočítá průměr hodnot na základě kritéria":
            print("AVERAGEIF: Nesprávně")
            correct = False
        else:
            print("AVERAGEIF: Správně")

        if correct:
            print("Všechny odpovědi jsou správné!")
        else:
            print("Zkuste to prosím znovu.")

q7_3_button.on_click(check_q7_3)

q7_3_grid = widgets.GridBox(
    children=[sumif_box, countif_box, averageif_box],
    layout=widgets.Layout(
        grid_template_columns="repeat(3, 1fr)",
        grid_gap="10px"
    )
)

display(q7_3_grid)
display(q7_3_button, q7_3_output)
```

```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

display(HTML("<h2>Interaktivní cvičení: Oprav chybu v Excel vzorcích</h2>"))
display(HTML("<p>V následujících otázkách jsou uvedeny Excel vzorce obsahující běžné chyby. Vaším úkolem je opravit každý vzorec a zadat správnou verzi do textového pole. Po stisknutí tlačítka získáte zpětnou vazbu.</p>"))

########################################
# Otázka 1: Chybějící závorka ve funkci
########################################
q1_label = widgets.HTML(
    value="<b>Otázka 1:</b> Opravte vzorec s chybějící závorkou ve funkci.<br>Chybně zapsaný vzorec: <code>=SUMA(A1:A5</code>"
)
q1_input = widgets.Text(
    value="",
    placeholder="Opravte vzorec...",
    description="Oprava:",
    layout=widgets.Layout(width='100%')
)
q1_button = widgets.Button(description="Zkontrolovat", button_style="info")
q1_output = widgets.Output()

expected_q1 = "=SUMA(A1:A5)"

def check_q1(b):
    with q1_output:
        clear_output()
        # Porovnání bez mezer
        if q1_input.value.replace(" ", "") == expected_q1.replace(" ", ""):
            print("Správně!")
        else:
            print("Nesprávně, zkuste to znovu.")

q1_button.on_click(check_q1)
display(q1_label, q1_input, q1_button, q1_output)

########################################
# Otázka 2: Špatný oddělovač argumentů
########################################
q2_label = widgets.HTML(
    value="<b>Otázka 2:</b> Opravte vzorec se špatným oddělovačem argumentů.<br>Chybně zapsaný vzorec: <code>=SUMA(A1;B1)</code>"
)
q2_input = widgets.Text(
    value="",
    placeholder="Opravte vzorec...",
    description="Oprava:",
    layout=widgets.Layout(width='100%')
)
q2_button = widgets.Button(description="Zkontrolovat", button_style="info")
q2_output = widgets.Output()

expected_q2 = "=SUMA(A1:B1)"

def check_q2(b):
    with q2_output:
        clear_output()
        if q2_input.value.replace(" ", "") == expected_q2.replace(" ", ""):
            print("Správně!")
        else:
            print("Nesprávně, zkuste to znovu.")

q2_button.on_click(check_q2)
display(q2_label, q2_input, q2_button, q2_output)

########################################
# Otázka 3: Oprava relativních/absolutních odkazů
########################################
q3_label = widgets.HTML(
    value="<b>Otázka 3:</b> Opravte vzorec tak, aby odkaz na buňku B1 byl absolutní.<br>Chybně zapsaný vzorec: <code>=A1*B1</code>"
)
q3_input = widgets.Text(
    value="",
    placeholder="Opravte vzorec...",
    description="Oprava:",
    layout=widgets.Layout(width='100%')
)
q3_button = widgets.Button(description="Zkontrolovat", button_style="info")
q3_output = widgets.Output()

expected_q3 = "=A1*$B$1"

def check_q3(b):
    with q3_output:
        clear_output()
        if q3_input.value.replace(" ", "") == expected_q3.replace(" ", ""):
            print("Správně!")
        else:
            print("Nesprávně, zkuste to znovu.")

q3_button.on_click(check_q3)
display(q3_label, q3_input, q3_button, q3_output)
```
