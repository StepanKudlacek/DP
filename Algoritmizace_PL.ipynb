```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output
import random

# Z√°kladn√≠ nastaven√≠
grid_size = 8
robot_x, robot_y = 0, 0
cil_x, cil_y = grid_size - 1, grid_size - 1
prekazky = []

# Vytvo≈ôen√≠ v√Ωstupn√≠ho widgetu pro hern√≠ plochu
output = widgets.Output()

# Vytvo≈ôen√≠ tlaƒç√≠tek
up_btn = widgets.Button(description="Nahoru")
down_btn = widgets.Button(description="Dol≈Ø")
left_btn = widgets.Button(description="Vlevo")
right_btn = widgets.Button(description="Vpravo")
reset_btn = widgets.Button(description="Reset")

# Funkce pro zobrazen√≠ hern√≠ plochy
def zobraz_plochu():
    with output:
        clear_output(wait=True)
        # Vytvo≈ôen√≠ textov√© reprezentace hern√≠ plochy
        for y in range(grid_size):
            row = ""
            for x in range(grid_size):
                if x == robot_x and y == robot_y:
                    row += "R "  # Robot
                elif x == cil_x and y == cil_y:
                    row += "C "  # C√≠l
                elif (x, y) in prekazky:
                    row += "X "  # P≈ôek√°≈æka
                else:
                    row += ". "  # Pr√°zdn√© pole
            print(row)

        # Kontrola c√≠le
        if robot_x == cil_x and robot_y == cil_y:
            print("\nGratulujeme! Robot dos√°hl c√≠le!")

# Funkce pro ovƒõ≈ôen√≠, zda existuje cesta k c√≠li
def existuje_cesta():
    # Jednoduch√° implementace BFS (Breadth-First Search) pro kontrolu dostupnosti c√≠le
    navstiveno = set()
    fronta = [(robot_x, robot_y)]

    while fronta:
        x, y = fronta.pop(0)
        if (x, y) == (cil_x, cil_y):
            return True

        if (x, y) in navstiveno:
            continue

        navstiveno.add((x, y))

        # Zkontrolujeme v≈°echny ƒçty≈ôi smƒõry
        for dx, dy in [(0, -1), (0, 1), (-1, 0), (1, 0)]:
            nx, ny = x + dx, y + dy
            if (0 <= nx < grid_size and 0 <= ny < grid_size and
                (nx, ny) not in prekazky and (nx, ny) not in navstiveno):
                fronta.append((nx, ny))

    return False

# Funkce pro inicializaci hry
def init_game():
    global robot_x, robot_y, prekazky

    # Resetov√°n√≠ pozic
    robot_x, robot_y = 0, 0

    attempt_count = 0
    while True:
        attempt_count += 1
        prekazky = []

        # Rovnomƒõrnƒõ rozlo≈æ√≠me p≈ôek√°≈æky po cel√© mapƒõ

        # 1. Nejprve vytvo≈ô√≠me m≈ô√≠≈ækov√Ω vzor p≈ôek√°≈æek
        for x in range(1, grid_size, 2):
            for y in range(1, grid_size, 2):
                if (x, y) != (cil_x, cil_y) and random.random() < 0.7:  # 70% ≈°ance
                    prekazky.append((x, y))

        # 2. P≈ôid√°me konkr√©tn√≠ p≈ôek√°≈æky na kl√≠ƒçov√Ωch m√≠stech
        key_positions = [
            # Prvn√≠ a druh√Ω ≈ô√°dek
            (1, 0), (3, 0), (5, 0),
            (2, 1), (4, 1), (6, 1),

            # Prvn√≠ a druh√Ω sloupec
            (0, 1), (0, 3), (0, 5),
            (1, 2), (1, 4), (1, 6),

            # St≈ôedn√≠ oblast
            (2, 2), (2, 4), (2, 6),
            (4, 2), (4, 4), (4, 6),
            (6, 2), (6, 4), (6, 6),

            # Oblasti bl√≠zko c√≠le
            (grid_size-2, grid_size-2),
            (grid_size-2, grid_size-3),
            (grid_size-3, grid_size-2),
            (grid_size-3, grid_size-3),

            # Cesta doprava dole
            (grid_size-2, grid_size-1),
            (grid_size-3, grid_size-1),
            (grid_size-4, grid_size-1),
            (grid_size-1, grid_size-2),
            (grid_size-1, grid_size-3),
            (grid_size-1, grid_size-4)
        ]

        for pos in key_positions:
            if pos not in prekazky and pos != (robot_x, robot_y) and pos != (cil_x, cil_y) and random.random() < 0.6:  # 60% ≈°ance
                prekazky.append(pos)

        # 3. P≈ôid√°me n√°hodn√© p≈ôek√°≈æky
        for _ in range(10):
            x, y = random.randint(0, grid_size-1), random.randint(0, grid_size-1)
            if (x, y) != (robot_x, robot_y) and (x, y) != (cil_x, cil_y) and (x, y) not in prekazky:
                prekazky.append((x, y))

        # 4. Ujist√≠me se, ≈æe cesta kolem startu a c√≠le je trochu volnƒõj≈°√≠
        if (1, 0) in prekazky and (0, 1) in prekazky:
            if random.random() < 0.5:
                prekazky.remove((1, 0))
            else:
                prekazky.remove((0, 1))

        if (grid_size-2, grid_size-1) in prekazky and (grid_size-1, grid_size-2) in prekazky:
            if random.random() < 0.5:
                prekazky.remove((grid_size-2, grid_size-1))
            else:
                prekazky.remove((grid_size-1, grid_size-2))

        # Ovƒõ≈ô√≠me, ≈æe existuje cesta k c√≠li
        if existuje_cesta() and len(prekazky) >= 15:
            break

        # Pokud jsme ji≈æ udƒõlali mnoho pokus≈Ø, uvoln√≠me po≈æadavky
        if attempt_count > 20 and existuje_cesta():
            break

    zobraz_plochu()

# Funkce pro pohyb
def move_up(b):
    global robot_y
    if robot_y > 0 and (robot_x, robot_y - 1) not in prekazky:
        robot_y -= 1
        zobraz_plochu()

def move_down(b):
    global robot_y
    if robot_y < grid_size - 1 and (robot_x, robot_y + 1) not in prekazky:
        robot_y += 1
        zobraz_plochu()

def move_left(b):
    global robot_x
    if robot_x > 0 and (robot_x - 1, robot_y) not in prekazky:
        robot_x -= 1
        zobraz_plochu()

def move_right(b):
    global robot_x
    if robot_x < grid_size - 1 and (robot_x + 1, robot_y) not in prekazky:
        robot_x += 1
        zobraz_plochu()

def reset(b):
    init_game()

# P≈ôi≈ôazen√≠ funkc√≠ k tlaƒç√≠tk≈Øm
up_btn.on_click(move_up)
down_btn.on_click(move_down)
left_btn.on_click(move_left)
right_btn.on_click(move_right)
reset_btn.on_click(reset)

# Vytvo≈ôen√≠ layoutu ovl√°dac√≠ch prvk≈Ø
controls = widgets.VBox([
    widgets.HBox([widgets.Label(layout=widgets.Layout(width='20px')), up_btn, widgets.Label(layout=widgets.Layout(width='20px'))]),
    widgets.HBox([left_btn, down_btn, right_btn]),
    reset_btn
])

# Vytvo≈ôen√≠ celkov√©ho layoutu
layout = widgets.VBox([
    widgets.HTML("<h2>Robot Simulator</h2>"),
    widgets.HTML("<p>Dosta≈àte robota (R) k c√≠li (C) pomoc√≠ tlaƒç√≠tek. P≈ôek√°≈æky jsou v≈°ude!</p>"),
    output,
    controls
])

# Inicializace a zobrazen√≠ hry
init_game()
display(layout)
```

```python
# @title
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output
import random

# Definice zjednodu≈°en√Ωch krok≈Ø algoritmu K√°men-N≈Ø≈æky-Pap√≠r
kroky_algoritmu = [
    "Zat≈ôesen√≠ micro:bitem spust√≠ hru",
    "Zobrazit na displeji odpoƒçet 3, 2, 1",
    "Vygenerovat n√°hodn√© ƒç√≠slo od 1 do 3 pro volbu poƒç√≠taƒçe",
    "Podm√≠nka: Kdy≈æ n√°hodn√© ƒç√≠slo = 1, zobrazit symbol K√°men",
    "Podm√≠nka: Kdy≈æ n√°hodn√© ƒç√≠slo = 2, zobrazit symbol Pap√≠r",
    "Podm√≠nka: Kdy≈æ n√°hodn√© ƒç√≠slo = 3, zobrazit symbol N≈Ø≈æky",
    "Poƒçkat 3 sekundy, aby hr√°ƒç vidƒõl v√Ωsledek",
    "Vr√°tit se na zaƒç√°tek a ƒçekat na dal≈°√≠ zat≈ôesen√≠"
]

# Zam√≠ch√°n√≠ krok≈Ø pro v√Ωchoz√≠ stav
kroky_zamichane = kroky_algoritmu.copy()
random.shuffle(kroky_zamichane)

# Nadpis aktivity
nadpis = widgets.HTML(
    value='<h2>Algoritmus hry K√°men-N≈Ø≈æky-Pap√≠r pro micro:bit</h2>'
    '<p>Se≈ôaƒète kroky algoritmu do spr√°vn√©ho po≈ôad√≠. Kliknƒõte na krok a pomoc√≠ tlaƒç√≠tek "Nahoru ‚Üë" a "Dol≈Ø ‚Üì" ho p≈ôesu≈àte.</p>'
)

# Widget pro zobrazen√≠ seznamu krok≈Ø s mo≈ænost√≠ ≈ôazen√≠
kroky_selectmultiple = widgets.SelectMultiple(
    options=kroky_zamichane,
    rows=9,  # Zobraz√≠me v≈°echny ≈ô√°dky najednou
    description='Kroky:',
    layout=widgets.Layout(width='80%', height='auto'),
    disabled=False
)

# Tlaƒç√≠tka pro pohyb polo≈æek nahoru a dol≈Ø
btn_nahoru = widgets.Button(
    description='Nahoru ‚Üë',
    layout=widgets.Layout(width='100px'),
    button_style='primary'
)
btn_dolu = widgets.Button(
    description='Dol≈Ø ‚Üì',
    layout=widgets.Layout(width='100px'),
    button_style='primary'
)

# Tlaƒç√≠tko pro kontrolu ≈ôe≈°en√≠
btn_kontrola = widgets.Button(
    description='Zkontrolovat ≈ôe≈°en√≠',
    button_style='success',
    layout=widgets.Layout(width='150px')
)

# Tlaƒç√≠tko pro zam√≠ch√°n√≠ krok≈Ø
btn_zamichat = widgets.Button(
    description='Zam√≠chat kroky',
    button_style='warning',
    layout=widgets.Layout(width='150px')
)

# V√Ωstupn√≠ oblast pro zpƒõtnou vazbu
vystup = widgets.Output()

# Pomocn√© funkce
def posunout_nahoru(b):
    """Posune vybran√© kroky o jedno m√≠sto nahoru"""
    if not kroky_selectmultiple.index:
        return

    # Z√≠sk√°n√≠ index≈Ø vybran√Ωch polo≈æek
    indexy = list(kroky_selectmultiple.index)

    # Nem≈Ø≈æeme posunout nahoru prvn√≠ polo≈æku
    if 0 in indexy:
        return

    # Z√≠sk√°n√≠ aktu√°ln√≠ch mo≈ænost√≠
    moznosti = list(kroky_selectmultiple.options)

    # Posun vybran√Ωch polo≈æek nahoru
    for i in sorted(indexy):
        moznosti[i-1], moznosti[i] = moznosti[i], moznosti[i-1]

    # Aktualizace seznamu
    kroky_selectmultiple.options = moznosti

    # Aktualizace v√Ωbƒõru
    kroky_selectmultiple.index = tuple(i-1 for i in indexy)

def posunout_dolu(b):
    """Posune vybran√© kroky o jedno m√≠sto dol≈Ø"""
    if not kroky_selectmultiple.index:
        return

    # Z√≠sk√°n√≠ index≈Ø vybran√Ωch polo≈æek
    indexy = list(kroky_selectmultiple.index)

    # Nem≈Ø≈æeme posunout dol≈Ø posledn√≠ polo≈æku
    if len(kroky_selectmultiple.options) - 1 in indexy:
        return

    # Z√≠sk√°n√≠ aktu√°ln√≠ch mo≈ænost√≠
    moznosti = list(kroky_selectmultiple.options)

    # Posun vybran√Ωch polo≈æek dol≈Ø (zaƒç√≠n√°me od konce, abychom nenaru≈°ili indexy)
    for i in sorted(indexy, reverse=True):
        moznosti[i], moznosti[i+1] = moznosti[i+1], moznosti[i]

    # Aktualizace seznamu
    kroky_selectmultiple.options = moznosti

    # Aktualizace v√Ωbƒõru
    kroky_selectmultiple.index = tuple(i+1 for i in indexy)

def zkontrolovat_reseni(b):
    """Zkontroluje, zda jsou kroky se≈ôazeny spr√°vnƒõ"""
    with vystup:
        vystup.clear_output()

        aktualni_poradi = list(kroky_selectmultiple.options)

        # Porovn√°n√≠ se spr√°vn√Ωm po≈ôad√≠m
        spravne = True
        nespravne_kroky = []

        for i, (aktualni, spravny) in enumerate(zip(aktualni_poradi, kroky_algoritmu)):
            if aktualni != spravny:
                spravne = False
                nespravne_kroky.append(i+1)

        if spravne:
            display(HTML("<div style='background-color: #dff0d8; padding: 15px; border-radius: 5px; margin: 10px 0; font-weight: bold;'>Skvƒõl√° pr√°ce! üéâ Algoritmus je se≈ôazen spr√°vnƒõ.</div>"))
        else:
            # Sestaven√≠ zpr√°vy o chybn√Ωch kroc√≠ch
            chybne = ', '.join(str(krok) for krok in nespravne_kroky)
            display(HTML(f"<div style='background-color: #f2dede; padding: 15px; border-radius: 5px; margin: 10px 0;'><b>Algoritmus nen√≠ spr√°vnƒõ se≈ôazen.</b><br>Kroky, kter√© nejsou na spr√°vn√© pozici: {chybne}<br>Zkus to znovu!</div>"))

def zamichat_kroky(b):
    """Zam√≠ch√° kroky v seznamu"""
    moznosti = list(kroky_selectmultiple.options)
    random.shuffle(moznosti)
    kroky_selectmultiple.options = moznosti
    with vystup:
        vystup.clear_output()
        display(HTML("<div style='background-color: #d9edf7; padding: 15px; border-radius: 5px; margin: 10px 0;'>Kroky byly zam√≠ch√°ny. Se≈ôaƒè je do spr√°vn√©ho po≈ôad√≠!</div>"))

# P≈ôi≈ôazen√≠ akc√≠ k tlaƒç√≠tk≈Øm
btn_nahoru.on_click(posunout_nahoru)
btn_dolu.on_click(posunout_dolu)
btn_kontrola.on_click(zkontrolovat_reseni)
btn_zamichat.on_click(zamichat_kroky)

# Instrukce box - zjednodu≈°en√© tipy
instrukce_box = widgets.HTML(
    value="""
    <div style="background-color: #f8f9fa; padding: 15px; border-radius: 5px; margin: 10px 0;">
        <h4>Jak to funguje:</h4>
        <p>Micro:bit v t√©to h≈ôe pouze n√°hodnƒõ vyb√≠r√° a zobrazuje jeden ze t≈ô√≠ symbol≈Ø (K√°men, Pap√≠r nebo N≈Ø≈æky)
        na z√°kladƒõ n√°hodnƒõ vygenerovan√©ho ƒç√≠sla.</p>
    </div>
    """
)

# Vytvo≈ôen√≠ layoutu
tlacitka_pohyb = widgets.VBox([btn_nahoru, btn_dolu], layout=widgets.Layout(align_items='center', padding='10px'))
tlacitka_akce = widgets.HBox([btn_kontrola, btn_zamichat], layout=widgets.Layout(justify_content='center', padding='10px'))
layout = widgets.VBox([
    nadpis,
    instrukce_box,
    widgets.HBox([kroky_selectmultiple, tlacitka_pohyb], layout=widgets.Layout(align_items='center')),
    tlacitka_akce,
    vystup
], layout=widgets.Layout(width='100%'))

# Zobrazen√≠ widgetu
display(layout)
```

```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

# Vytvo≈ôen√≠ nadpisu a popisu √∫kolu
nadpis = widgets.HTML(
    value="""
    <h2>Dopl≈àte chybƒõj√≠c√≠ matematick√© operace</h2>
    <p>
        V k√≥du n√≠≈æe je implementov√°no pouze sƒç√≠t√°n√≠. Va≈°√≠m √∫kolem je doplnit zb√Ωvaj√≠c√≠ operace:
        <ul>
            <li>Odƒç√≠t√°n√≠ (-)</li>
            <li>N√°soben√≠ (*)</li>
            <li>Dƒõlen√≠ (/)</li>
        </ul>
    </p>
    """
)

# Editor k√≥du pro doplnƒõn√≠ chybƒõj≈°√≠ch operac√≠
kod_template = """# Funkce pro v√Ωpoƒçet
def vypocitej(b):
    with vystup:
        clear_output()
        a = cislo1.value
        b = cislo2.value
        op = operace.value

        if op == '+':
            print(f"{a} + {b} = {a + b}")
        # DOPL≈áTE CHYBƒöJ√çC√ç OPERACE
        elif op == '-':
            # Zde dopi≈°te k√≥d pro odƒç√≠t√°n√≠
        elif op == '*':
            # Zde dopi≈°te k√≥d pro n√°soben√≠
        elif op == '/':
            # Zde dopi≈°te k√≥d pro dƒõlen√≠"""

kod_editor = widgets.Textarea(
    value=kod_template,
    placeholder='Dopl≈àte k√≥d zde...',
    description='K√≥d:',
    disabled=False,
    layout=widgets.Layout(width='100%', height='220px')
)

# Testovac√≠ prost≈ôed√≠ pro vyzkou≈°en√≠ k√≥du
cislo1 = widgets.IntSlider(description="Prvn√≠ ƒç√≠slo:", min=-10, max=10, step=1, value=5)
cislo2 = widgets.IntSlider(description="Druh√© ƒç√≠slo:", min=-10, max=10, step=1, value=3)
operace = widgets.Dropdown(
    options=['+', '-', '*', '/'],
    value='+',
    description='Operace:'
)
tlacitko_test = widgets.Button(
    description="Vyzkou≈°et k√≥d",
    button_style='info',
    layout=widgets.Layout(margin='10px 0px')
)
vystup_test = widgets.Output()

# Tlaƒç√≠tko pro kontrolu
tlacitko_kontrola = widgets.Button(
    description="Zkontrolovat ≈ôe≈°en√≠",
    button_style='success',
    layout=widgets.Layout(margin='10px 0px')
)
vystup_kontrola = widgets.Output()

# Funkce pro testov√°n√≠ k√≥du
def testuj_kod(b):
    try:
        kod = kod_editor.value

        # Vytvo≈ô√≠me kopii promƒõnn√Ωch pro bezpeƒçn√© vyhodnocen√≠
        test_locals = {
            'cislo1': type('', (), {'value': cislo1.value})(),
            'cislo2': type('', (), {'value': cislo2.value})(),
            'operace': type('', (), {'value': operace.value})(),
            'vystup': vystup_test,
            'clear_output': clear_output,
            'print': print
        }

        # Vyhodnot√≠me k√≥d
        with vystup_test:
            clear_output()
            exec(kod, globals(), test_locals)

            # Zavol√°me funkci vypocitej
            if 'vypocitej' in test_locals:
                test_locals['vypocitej'](None)
            else:
                print("Chyba: Funkce 'vypocitej' nebyla definov√°na!")

    except Exception as e:
        with vystup_test:
            clear_output()
            print(f"Chyba v k√≥du: {str(e)}")

# Kontrola ≈ôe≈°en√≠ - detailn√≠
def zkontroluj_reseni(b):
    kod = kod_editor.value

    # P≈ôesn√© vzory, kter√© mus√≠ b√Ωt v k√≥du
    kontroly = [
        ('-', "Chyb√≠ implementace odƒç√≠t√°n√≠", "{a} - {b} = {a - b}"),
        ('*', "Chyb√≠ implementace n√°soben√≠", "{a} * {b} = {a * b}"),
        ('/', "Chyb√≠ implementace dƒõlen√≠", "{a} / {b} = {a / b}")
    ]

    # V√Ωsledky
    with vystup_kontrola:
        clear_output()

        html = "<div style='background-color: #f8f9fa; padding: 15px; border-radius: 5px;'>"
        html += "<h3>Kontrola k√≥du:</h3><ul>"

        # P≈ô√≠znak kompletn√≠ho ≈ôe≈°en√≠
        vse_splneno = True

        for operace, zprava, pattern in kontroly:
            # Poƒçet v√Ωskyt≈Ø dan√©ho vzoru
            pocet_vyskytu = kod.count(pattern)

            if pocet_vyskytu > 0:
                html += f"<li style='color: green;'>‚úì {zprava}</li>"
            else:
                html += f"<li style='color: red;'>‚úó {zprava}</li>"
                vse_splneno = False

        html += "</ul>"

        # Celkov√© hodnocen√≠
        if vse_splneno:
            html += "<div style='background-color: #dff0d8; padding: 10px; border-radius: 5px; margin-top: 10px;'><b>Skvƒõl√° pr√°ce!</b> V≈°echny operace jsou spr√°vnƒõ implementov√°ny.</div>"
        else:
            html += "<div style='background-color: #f2dede; padding: 10px; border-radius: 5px; margin-top: 10px;'><b>Nƒõco je≈°tƒõ chyb√≠.</b> Zkontrolujte implementaci v≈°ech operac√≠.</div>"

        html += "</div>"

        display(HTML(html))

# Zamy≈°len√≠ nad ≈ôe≈°en√≠m
zamysleni = widgets.HTML(
    value="""
    <div style="background-color: #d9edf7; padding: 15px; border-radius: 5px; margin-top: 10px;">
        <h4>Zamy≈°len√≠ nad ≈ôe≈°en√≠m:</h4>
        <p>Nyn√≠ se zamyslete nad n√°sleduj√≠c√≠mi ot√°zkami:</p>
        <ol>
            <li>Co by se mohlo st√°t, kdybychom zkusili dƒõlit nulou?</li>
            <li>Jak byste mohli zabr√°nit chyb√°m p≈ôi matematick√Ωch operac√≠ch?</li>
            <li>Jak√© dal≈°√≠ kontroly byste mohli p≈ôidat do va≈°eho k√≥du?</li>
        </ol>
        <p><b>Tip:</b> Zkuste vymyslet zp≈Øsob, jak ochr√°nit v√°≈° program p≈ôed neoƒçek√°van√Ωmi vstupy.</p>
    </div>
    """
)

# P≈ôi≈ôazen√≠ funkc√≠ k tlaƒç√≠tk≈Øm
tlacitko_test.on_click(testuj_kod)
tlacitko_kontrola.on_click(zkontroluj_reseni)

# Vytvo≈ôen√≠ layoutu
layout = widgets.VBox([
    nadpis,
    kod_editor,
    widgets.HBox([tlacitko_test, tlacitko_kontrola]),
    vystup_kontrola,
    zamysleni,
    vystup_test
])

# Zobrazen√≠ widget≈Ø
display(layout)
```
